# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
ArithmeticError is simply an error that occurs during numeric calculations.

ArithmeticError types in Python include:

1. OverFlowError
2. ZeroDivisionError
3. FloatingPointError
```

```python
 #ZeroDivisionError
try:
  arithmetic = 5/0
  print(arithmetic)
except ArithmeticError:
  print('You have just made an Arithmetic error')

```

```python
# OverFlowErro
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened")
    print(f"{e}, {e.__class__}")
```

 # Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

1. LookupError Exception is the Base class for errors raised when something can’t be found. 
2. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError.
3. An IndexError is raised when a sequence reference is out of range.

```python
# Key Error
subjects = {'Sree': 'Maths',
            'Ram': 'Biology',
            'Shyam': 'Science',
            'Abdul': 'Hindi'}
# try block
try:
    print('subject of Fharan is:',
          subjects['Fharan'])
# except Block
except KeyError:
    print("Fharan's records doesn't exist")
```

```python
# Index Error
import sys
print("Lets Try to find out the result for students in the class")
n=int(input())
try:
    marks_list = [20,30,50,80,90]
    print("Marks for roll no " + str(n) +" is " +str(marks_list[n]))
except IndexError as e:
    print("Student with roll no "+str(n)+" does not exist")
    print(e)
```

# Q6. List down some best practices for exception handling in python.

```python
# 1. Use always a specific exceptions

try :
    10/0
except ZeroDivisionError as e :
    print(e)
```

```python
# 2. Print always valid msg

try :
    10/0
except ZeroDivisionError as e :
    print("This is my Zerodivision error i am handling", e)
```

```python
# 3. Always try to log

import logging

logging.basicConfig(filename = "error.log", level = logging.ERROR)
try :
    10/0
except ZeroDivisionError as e :
    logging.error("This is my Zerodivision error i am handling {} ".format(e))
```

```python
# 4. Always avoid to writr multiple exception handling 

logging.basicConfig(filename = "error.log", level = logging.ERROR)
try :
    10/0
except FileNotFoundError as e :
    logging.error("This is my file not fond error i am handling {} ".format(e))
except AttributeError as e :
    logging.error("This is my attribute error i am handling {} ".format(e))
except ZeroDivisionError as e :
    logging.error("This is my Zerodivision error i am handling {} ".format(e))
```

```python
# 5. Cleanup all the resources

try :
    with open("test.txt", 'w') as f:
        f.write("This is my msg")
        
except FileNotFoundError as e :
    logging.error("This is my file not fond error i am handling {} ".format(e))
    
finally :
    f.close()
```
